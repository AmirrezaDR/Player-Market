```python
df=df.rename(columns={'base price (in lacs)':'base_price','final price (in lacs)':'final_price'})
```

# Replacing the NAN with the mean of there column

```python

df['base_price']=df['base_price'].fillna(df['base_price'].mean())
df['final_price']=df['final_price'].fillna(df['final_price'].mean())
df.duplicated().sum()
```

```python
df.shape
```

```python
# retained_df=df.loc[(df['status'] == "RETAINED")]
# country_group = retained_df.groupby('nationality')
country_group = df.groupby('nationality')

for nationality, group in country_group:
    print(f"Nationality: {nationality}")
    print(group)
    print("\n")  # Print a newline for better separation
```

```python
india_data= country_group.get_group("India")
sb.barplot(india_data)

```

```python
# c_g_s =df.groupby("nationality")[['final_price','base_price']].sum().reset_index()
# c_g_s
# sb.barplot(c_g_s)
# Aggregate the data by nationality
country_prices = df.groupby('nationality')[['base_price', 'final_price']].sum().reset_index()

# Set up the bar chart
fig, ax = plt.subplots(figsize=(12, 8))

# Define the positions of the bars on the x-axis
x = np.arange(len(country_prices['nationality']))

# Define the bar width
bar_width = 0.35

# Plot bars for base_price and final_price
ax.bar(x - bar_width/2, country_prices['base_price'], width=bar_width, label='Base Price')
ax.bar(x + bar_width/2, country_prices['final_price'], width=bar_width, label='Final Price')

# Set labels and title
ax.set_xlabel('Country')
ax.set_ylabel('Price (in lacs)')
ax.set_title('Base Price vs Final Price by Country')
ax.set_xticks(x)
# ax.set_xticklabels(country_prices['nationality'])
ax.set_xticklabels(country_prices['nationality'], rotation=45)

# Add a legend
ax.legend()

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Assuming your data is in a DataFrame named `df`
# Aggregate the data by nationality
country_prices = df.groupby('nationality')[['base_price', 'final_price']].sum().reset_index()

# Melt the data to long format for seaborn
melted_data = pd.melt(country_prices, id_vars='nationality', value_vars=['base_price', 'final_price'],
                      var_name='Price Type', value_name='Price')

# Set up the bar plot
plt.figure(figsize=(12, 8))
sb.barplot(data=melted_data, x='nationality', y='Price', hue='Price Type')

# Customize the plot
plt.xlabel('Country')
plt.ylabel('Price (in lacs)')
plt.title('Base Price vs Final Price by Country')
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()
plt.show()
```
