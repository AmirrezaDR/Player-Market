```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sb

pd.set_option("display.precision", 2)
df= pd.read_csv('iplauction2023.csv')
df
```

```python
df['base price (in lacs)'].mean()
```

```python
df=df.rename(columns={
    'base price (in lacs)':'base_price',
    'final price (in lacs)':'final_price'
})
```

```python
df.plot()
plt.show()
```

```python
df.isna().count()
```

```python
df.replace({
    -99999 : np.nan,
    -88888 : np.nan, 
})
```

```python
df['base_price'].fillna(df['base_price'].mean(),inplace= True)
df['final_price'].fillna(df['final_price'].mean(),inplace= True)
df.duplicated().sum()
```

```python
country_group = df.groupby('nationality')

for nationality, data in country_group:
    print(f"Nationality: {nationality}")
    print(data)
    print("\n")
```

```python
india_data= country_group.get_group("India")
sb.barplot(india_data)
```

```python
country_prices = df.groupby('nationality')[['base_price', 'final_price']].sum().reset_index()
melted_data = pd.melt(country_prices, id_vars='nationality', value_vars=['base_price', 'final_price'],var_name='Price Type', value_name='Price')
plt.figure(figsize=(12, 8))
sb.barplot(data=melted_data,x='nationality', y='Price', hue='Price Type')
plt.xlabel('Country')
plt.ylabel('Price (in lacs)')
plt.title('Base Price vs Final Price by Country')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
correletion = df['base_price'].corr(df.final_price )
print('correletion between the base price and final price is',correletion)
```

```python
RETAINED_df = df[(df['status']=='SOLD')&(df['nationality']=='India')]
RETAINED_df.plot()
plt.show()
```
